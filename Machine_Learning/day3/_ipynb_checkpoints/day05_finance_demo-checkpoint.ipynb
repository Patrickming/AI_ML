```python
import pandas as pd
```

```python
# 1、读取数据
data = pd.read_csv("./stock_day/stock_day.csv")
```

```python
data
```

```python
# 日期怎么处理？
data.index.values
```

```python
from mpl_finance import candlestick_ochl
```

```python
help(candlestick_ochl)
```

```python
# 准备quotes : sequence of (time, open, close, high, low, ...)
# 将字符串转换成日期
date = pd.to_datetime(data.index.values)
```

```python
from matplotlib.pylab import date2num
```

```python
data["time"] = date2num(date)
```

```python
day_k = data[["time", "open", "close", "high", "low"]]
```

```python
day_k = day_k.sort_index()
```

```python
day_k.head()
```

```python
# 准备axes
import matplotlib.pyplot as plt
```

```python
figure, axes = plt.subplots(nrows=1, ncols=1, figsize=(13, 10), dpi=100)

# 绘制日K线图
candlestick_ochl(axes, day_k.values[:50], width=0.4, colorup="r", colordown="g")

# 显示日期
axes.xaxis_date()

# 添加网格
plt.grid(linestyle="--", alpha=0.5)

plt.show()
```

### 绘制周K线图

```python
data.index = pd.to_datetime(data.index)
```

```python
data = data.sort_index()
```

```python
data
```

```python
# 数据重采样
week_k = data.resample("W").last()
```

```python
week_k.head()
```

```python
week_k["open"] = data["open"].resample("W").first()
```

```python
week_k["high"] = data["high"].resample("W").max()
```

```python
week_k["low"] = data["low"].resample("W").min()
```

```python
# sequence of (time, open, close, high, low, ...)
week_k = week_k[["time", "open", "close", "high", "low"]]
```

```python
figure, axes = plt.subplots(nrows=1, ncols=1, figsize=(13, 10), dpi=100)

# 绘制日K线图
candlestick_ochl(axes, week_k.values[:50], width=1.8, colorup="r", colordown="g")

# 显示日期
axes.xaxis_date()

# 添加网格
plt.grid(linestyle="--", alpha=0.5)

plt.show()
```
